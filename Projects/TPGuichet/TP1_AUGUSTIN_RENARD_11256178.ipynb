## Question 1 : Guichet libre-service - montant de la facture, arrondissement, détails du paiement (5,5 points)

Vous devez écrire un script python simulant le fonctionnement d'une caisse libre-service.

Le script demandera à l'utilisateur de saisir un montant monétaire à payer (c'est-à-dire le coût d'une facture) et arrondira ce montant à deux décimales ainsi qu'aux 5 ¢ les plus proches et l'affichera ensuite à l'utilisateur. Enfin, il faudrait s'assurer que le montant que l'utilisateur paiera est égal ou supérieur au montant de la facture due.

Pour cette question, vous devez réaliser les trois étapes suivantes :

1. obtenir de l'utilisateur le montant de la facture au cent près; il faut boucler tant que le montant n'est pas :
    - supérieur à 0.


2. arrondir le montant aux 5 ¢ près et l'afficher : c'est le montant qu'on doit payer lorsqu'on paie comptant :
    - de 0 ¢ à 2 ¢, on arrondi à 0 ¢; par exemple, __310,22 \$__ devient __310,20 \$__
    - de 3 ¢ à 7 ¢, on arrondi à 5 ¢; par exemple, __114,77 \$__ devient __114,75 \$__
    - de 8 ¢ à 9 ¢, on arrondi à 10 ¢; par exemple, __199,98 \$__ devient __200,00 \$__


3. s'assurer que le montant introduit pour le paiment est valide. Il faut boucler tant que le montant n'est pas :
    - supérieur ou égal au montant déterminé à l'étape 2.
    
_En utilisant `#`, expliquez votre code afin que nous puissions suivre votre logique._


### Trace d'exécution

#### Étape 1 : Génération et affichage du montant à payer



#### Étape 2 : Arrondissement et affichage du montant à payer au comptant



#### Étape 3 : Saisie du paiement



```python
# Étape 1 : Obtenir le montant de la facture
# Valeur initial du de la facture
client = 0

# On demande le montant de la facture et tant que le montant de la facture est inférieur ou égale à 0 alors ... 
while client < 0 or client == 0 : 
    client = input ("Quel est le montant de la facture ? ")
    client = float(client)
    
# Alors on continue de lui demander     
if (client < 0) : 
        print("Veuillez fournir un montant supérieur à 0")
elif (client == 0) : 
        print("Veuillez fournir un montant supérieur à 0")
        
# Une fois que c'est bon on affiche le montant total de la facture 
else : 
    print("Montant total de la facture: {0} $" .format(client))
```

```python
# Étape 2 : Arrondir le montant aux 5 centimes et l'afficher
# On calcul le quotient de la division euclidienne  pour isoler les centièmes
quotient = client//0.1

# On fait la meme chose pour isoler le reste et arrondir à 2 chiffres après la virgule
reste = round(client%0.1,2)

# On créer une variable pour la facture arrondi à la fin
facture_arrondi=0

# On créer une condition pour que le reste soit compris entre les différentes conditions et donc faire les ajustements directement
# sur la facture, puis on l'affiche une fois ajuster 
if reste>=0 and reste<=0.02 :
    facture_arrondi = client-reste
    
elif reste>=0.03 and reste<=0.07 :
    facture_arrondi = (client-reste)+0.05

else: 
    facture_arrondi = (client-reste)+0.1
    
facture_arrondi=round(facture_arrondi,2)    
print("Montant pour paiement en espèces: {:.2f}$" .format(facture_arrondi))
```

```python
# Étape 3 : Détails du paiement
# On demande la première valeur à entrer puis on lance le while si celle-ci est inferrieure à la valeur demandé précédemment
client_fin=0
client_fin = input("Veuillez introduire le total de vos billets et pièces: ")
client_fin = float(client_fin)

while (client_fin < facture_arrondi) :
    print("Veuillez fournir un montant supérieur ou égal à {:.2f}" .format(facture_arrondi))
    print("\"{:.2f}\" n'est pas un montant suffisant pour votre paiement." .format(client_fin))
    client_fin = input("Veuillez introduire le total de vos billets et pièces: ")
    client_fin = float(client_fin)

# On boucle jusqu'à ce que le montant entrer soit supérieur au montant de la facture_arrondi
    
print("Merci d'avoir soumis un montant suffisant pour votre paiement. ")
```

## Question 2 : Guichet libre-service - déterminer la monnaie à rendre (5,5 points)

Dépendamment du montant spécifié par l’utilisateur pour faire le paiement de la facture, le script calculera et affichera le montant de la monnaie à rendre incluant le nombre de billets et de pièces.

On supposera que le guichet est toujours approvisionné en billets et pièces (tous les billets et pièces existants en dollar canadien).

Pour cette sous-question, vous devez réaliser les deux étapes suivantes :

1. calculer et afficher le montant de monnaie à rendre.
2. calculer et afficher le nombre de billets et de pièces à rendre : vous devez afficher uniquement les dénominations de billets et de pièces dont le nombre requis est différent de 0. Votre guichet peut rendre les dénominations suivantes : 100 __\$__, 50 __\$__, 20 __\$__, 10 __\$__, 5 __\$__, 2 __\$__, 1 __\$__, 25 ¢, 10 ¢ et 5 ¢.

Pour que votre solution soit acceptée, la monnaie rendue doit toujours minimiser le nombre de billets et pièces au minimum.

_En utilisant `#`, expliquez votre code afin que nous puissions suivre votre logique._

### Trace d'exécution

#### Étape 1 : Calcul et affichage de la monnaie à rendre



#### Étape 2 : Calcul et affichage du nombre de billets et de pièces à rendre



```python
# Étape 1 : Calcul et affichage du montant de monnaie à rendre
# Rien de bien compliqué on demande et on affiche la différence entre le montant entré précédemment et le montant du paiement
monnaie = round (client_fin - facture_arrondi,1)
print ("Monnaie à rendre:{:.2f}$" .format(monnaie))
```

```python
# Étape 2 : Calcul et affichage du nombre de billets et de pièces à rendre 
# On créer deux premieres listes qui permetent d'enregistrer le nombre de billets et de pièces dont on va avoir à rendre
billets=[0,0,0,0,0]
pieces =[0,0,0,0,0]

# On créer deux autres listes pour les valeurs des billets et des pièces pour facilement les associers ensemble plus tard 
tableau_billets =[100,50,20,10,5]
tableau_pieces = [2,1,25,10,5]

print("Le guichet automatique déboursera :\n")

# Le while permet de calculer le nombre de billets et de pièces en regardant à chaque fois le montant restant après l'opération
# et de compter les différents billets et pièces que nous allons utilisés
while monnaie != 0 :
    monnaie = round(monnaie,2)
    if monnaie >= 100 :  
        monnaie -= 100
        billets[0] +=1
    elif monnaie >= 50 :  
        monnaie -= 50
        billets[1] +=1
    elif monnaie >= 20 :  
        monnaie -= 20
        billets[2] +=1
    elif monnaie >= 10 :  
        monnaie -= 10
        billets[3] +=1
    elif monnaie >= 5 :  
        monnaie -= 5
        billets[4] +=1
    elif monnaie >= 2 :  
        monnaie -= 2
        pieces[0] +=1
    elif monnaie >= 1 :  
        monnaie -= 1
        pieces[1] +=1
    elif monnaie >= 0.25 :  
        monnaie -= 0.25
        pieces[2] +=1
    elif monnaie >= 0.10 :  
        monnaie -= 0.10
        pieces[3] +=1
    elif monnaie >= 0.05 :  
        monnaie -= 0.05
        pieces[4] +=1

# Finalement on lance un for pour compter le nombre de billets dans un premier temps puis le nombre de pieces et de cents que 
# la machine va devoir rendre. Puis on affiche tout ca 
        
for i in range(len(billets)) :
    if billets[i] > 0 :
        print("{0} billets x {1}$" .format(billets[i], tableau_billets[i]))
for y in range(len(pieces)) : 
    if pieces[y] > 0 :
        if y < 2:
            print("{0} pièces x {1}$" .format(pieces[y], tableau_pieces[y]))
        else :
            print("{0} pièces x {1}¢" .format(pieces[y], tableau_pieces[y]))
```
