    - de 8   9 , on arrondi  10 ; par exemple, __199,98 \$__ devient __200,00 \$__

```python
# tape 1 : Obtenir le montant de la facture
# Valeur initial du de la facture
client = 0

# On demande le montant de la facture et tant que le montant de la facture est infrieur ou gale  0 alors ... 
while client < 0 or client == 0 : 
    client = input ("Quel est le montant de la facture ? ")
    client = float(client)
    
# Alors on continue de lui demander     
if (client < 0) : 
        print("Veuillez fournir un montant suprieur  0")
elif (client == 0) : 
        print("Veuillez fournir un montant suprieur  0")
        
# Une fois que c'est bon on affiche le montant total de la facture 
else : 
    print("Montant total de la facture: {0} $" .format(client))
```

```python
# tape 2 : Arrondir le montant aux 5 centimes et l'afficher
# On calcul le quotient de la division euclidienne  pour isoler les centimes
quotient = client//0.1

# On fait la meme chose pour isoler le reste et arrondir  2 chiffres aprs la virgule
reste = round(client%0.1,2)

# On crer une variable pour la facture arrondi  la fin
facture_arrondi=0

# On crer une condition pour que le reste soit compris entre les diffrentes conditions et donc faire les ajustements directement
# sur la facture, puis on l'affiche une fois ajuster 
if reste>=0 and reste<=0.02 :
    facture_arrondi = client-reste
    
elif reste>=0.03 and reste<=0.07 :
    facture_arrondi = (client-reste)+0.05

else: 
    facture_arrondi = (client-reste)+0.1
    
facture_arrondi=round(facture_arrondi,2)    
print("Montant pour paiement en espces: {:.2f}$" .format(facture_arrondi))
```

## Question 2 : Guichet libre-service - dterminer la monnaie  rendre (5,5 points)

```python
# tape 1 : Calcul et affichage du montant de monnaie  rendre
# Rien de bien compliqu on demande et on affiche la diffrence entre le montant entr prcdemment et le montant du paiement
monnaie = round (client_fin - facture_arrondi,1)
print ("Monnaie  rendre:{:.2f}$" .format(monnaie))
```

```python
# tape 2 : Calcul et affichage du nombre de billets et de pices  rendre 
# On crer deux premieres listes qui permetent d'enregistrer le nombre de billets et de pices dont on va avoir  rendre
billets=[0,0,0,0,0]
pieces =[0,0,0,0,0]

# On crer deux autres listes pour les valeurs des billets et des pices pour facilement les associers ensemble plus tard 
tableau_billets =[100,50,20,10,5]
tableau_pieces = [2,1,25,10,5]

print("Le guichet automatique dboursera :\n")

# Le while permet de calculer le nombre de billets et de pices en regardant  chaque fois le montant restant aprs l'opration
# et de compter les diffrents billets et pices que nous allons utiliss
while monnaie != 0 :
    monnaie = round(monnaie,2)
    if monnaie >= 100 :  
        monnaie -= 100
        billets[0] +=1
    elif monnaie >= 50 :  
        monnaie -= 50
        billets[1] +=1
    elif monnaie >= 20 :  
        monnaie -= 20
        billets[2] +=1
    elif monnaie >= 10 :  
        monnaie -= 10
        billets[3] +=1
    elif monnaie >= 5 :  
        monnaie -= 5
        billets[4] +=1
    elif monnaie >= 2 :  
        monnaie -= 2
        pieces[0] +=1
    elif monnaie >= 1 :  
        monnaie -= 1
        pieces[1] +=1
    elif monnaie >= 0.25 :  
        monnaie -= 0.25
        pieces[2] +=1
    elif monnaie >= 0.10 :  
        monnaie -= 0.10
        pieces[3] +=1
    elif monnaie >= 0.05 :  
        monnaie -= 0.05
        pieces[4] +=1

# Finalement on lance un for pour compter le nombre de billets dans un premier temps puis le nombre de pieces et de cents que 
# la machine va devoir rendre. Puis on affiche tout ca 
        
for i in range(len(billets)) :
    if billets[i] > 0 :
        print("{0} billets x {1}$" .format(billets[i], tableau_billets[i]))
for y in range(len(pieces)) : 
    if pieces[y] > 0 :
        if y < 2:
            print("{0} pices x {1}$" .format(pieces[y], tableau_pieces[y]))
        else :
            print("{0} pices x {1}" .format(pieces[y], tableau_pieces[y]))
```
